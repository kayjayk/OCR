```python
import json
import base64
import requests
import os
```

```python
root_dir = '../../../../../data/00_nh_poc/11_legal_advice_scan_149'
png_dir = os.path.join(root_dir, 'png_naver')
txt_dir = os.path.join(root_dir, 'txt_naver')
```

```python
for file_name in os.listdir(png_dir):
    with open(os.path.join(png_dir, file_name), "rb") as f:
        img = base64.b64encode(f.read())

    URL = "https://39a9eeb4312a4311ae9dd849104d12cb.apigw.ntruss.com/custom/v1/2460/d72bb34bf7d8464531ec1fe67e44c3205d5c45e52ba692743018bec955a7a65e/general"
    KEY = "bUtVbXN1QU1zVFp2Y2JLemVnbHpTTG5XZG5naHZtanc="
    headers = {
        "Content-Type": "application/json",
        "X-OCR-SECRET": KEY
    }
    data = {
        "version": "V1",
        "requestId": "kayjayk", # 요청을 구분하기 위한 ID, 사용자가 정의
        "timestamp": 0, # 현재 시간값
        "images": [
            {
                "name": "sample_image",
                "format": "png",
                "data": img.decode('utf-8')
            }
        ]
    }

    data = json.dumps(data)
    response = requests.post(URL, data=data, headers=headers)
    res = json.loads(response.text)

    result = []
    for i in range(len(res['images'][0]['fields'])):
        result.append(res['images'][0]['fields'][i]['inferText'])

    text = " ".join(result)
    file_name_wo_ext = file_name.split('.')[0]

    text_file = open(os.path.join(txt_dir, file_name_wo_ext) + '.txt', 'w')
    text_file.write(text)
    text_file.close()
```
